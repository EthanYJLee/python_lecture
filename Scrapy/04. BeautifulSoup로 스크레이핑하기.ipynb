```python
from bs4 import BeautifulSoup
import urllib.request as req

url = "https://zeushahn.github.io/Test/python/bs_exam01.html"

# res = 소스 가져오기
res = req.urlopen(url)
print(res)

# BeautifulSoup으로 분석하기
soup = BeautifulSoup(res,"html.parser")
print(soup)
```

```python
# 원하는 부분 추출하기 (스크레이핑이란?)
h1 = soup.html.body.h1
# 요소의 글자 출력하기 : String, text
print("h1 =",h1.string)
```

```python
# 원하는 부분 추출하기 (웹 페이지를 분석하는 것)
p1 = soup.html.body.p
# 요소의 글자 출력하기 : String, text
print("p1 =",p1.text)

# 다음번째 p 가져올 때:
# </p> (닫는 태그도 카운트하기 때문에 next_sibling 두 번 사용한다.)
p2 = p1.next_sibling.next_sibling
# 요소의 글자 출력하기 : String, text
print("p2 =",p2.text)
```

### CSS 선택자 사용하기
- BeautifulSoup은 JavaScript Library인 JQuery처럼 CSS 선택자를 지정해서 원하는 요소를 추출 가능.

```python
from bs4 import BeautifulSoup
import urllib.request as req
url = "https://zeushahn.github.io/Test/python/bs_exam04.html"

# res = 소스 가져오기
res = req.urlopen(url)


# BeautifulSoup으로 분석하기
soup = BeautifulSoup(res,"html.parser")

print(soup)
```

### 필요한 부분을 CSS Query로 추출하기 (#: id, .: class, >: 자식, 빈칸: 후손)

```python
# id가 meigen인 div의 하위에 있는 (자식인) h1
h1 = soup.select_one("div#meigen > h1").string
print("h1 =",h1)

# id가 meigen인 div의 하위에 있는 ul 클래스 중 id가 items인 ul을 찾고 그 하위의 li를 전부 출력
li_list = soup.select("div#meigen > ul.items > li")
for li in li_list:
    print("li =",li.string)
```

```python
# 미국, 일본, 유럽연합, 중국 환율
#exchangeList > li.on > a.head.usd > div > span.value
#exchangeList > li.on > a.head.jpy > div > span.value
#exchangeList > li.on > a.head.eur > div > span.value
#exchangeList > li.on > a.head.cny > div > span.value

from bs4 import BeautifulSoup
import urllib.request as req
url = "https://finance.naver.com/marketindex/"

# res = 소스 가져오기
res = req.urlopen(url)

# BeautifulSoup으로 분석하기
soup = BeautifulSoup(res,"html.parser")
exchange = soup.select(".value")
country = ['미국','일본','유럽연합','중국']
i = 0
for ex in exchange:
    print(country[i],'\t:',ex.string.rjust(10))
    i += 1
    if i >= len(country):
        break
```

### 다음영화 연간 순위 가져오기

```python
#mainContent > div > div.box_boxoffice > ol > li:nth-child(1) > div > div.thumb_cont > strong > a
#mainContent > div > div.box_boxoffice > ol > li:nth-child(50) > div > div.thumb_cont > strong > a
#mainContent > div > div.box_boxoffice > ol > li:nth-child(1) > div > div.thumb_cont > strong > a
from bs4 import BeautifulSoup
import urllib.request as req
import time

# url = "https://movie.daum.net/ranking/boxoffice/yearly"
# url = "https://movie.daum.net/ranking/boxoffice/yearly?date=2021"

# 년도 받아오기
prompt = """
1. 올해 순위 보기
2. 예전 순위 보기
0. 나가기
"""
num = 10
while num != 0:
    print(prompt)
    num = int(input())
    if num == 1:
        url = "https://movie.daum.net/ranking/boxoffice/yearly"
        # res = 소스 가져오기
        res = req.urlopen(url)

        # BeautifulSoup으로 분석하기
        soup = BeautifulSoup(res,"html.parser")
        i = list(range(1,11))
        for i in i:
            time.sleep(0.1)
            rank = soup.select_one("#mainContent > div > div.box_boxoffice > ol > li:nth-child(%s) > div > div.thumb_cont > strong > a"%i)
            print(i,'위:\t',rank.text)
        continue
     
    elif num == 2:
        print('년도를 입력하세요')
        year = int(input())
        if year in range(2004,2022):
            url = "https://movie.daum.net/ranking/boxoffice/yearly?date=%d"%year

            # res = 소스 가져오기
            res = req.urlopen(url)

            # BeautifulSoup으로 분석하기
            soup = BeautifulSoup(res,"html.parser")
            i = list(range(1,11))
            for i in i:
                time.sleep(0.1)
                rank = soup.select_one("#mainContent > div > div.box_boxoffice > ol > li:nth-child(%s) > div > div.thumb_cont > strong > a"%i)
                print(i,'위:\t',rank.text)
            continue
        else:
            break
    
        
    elif num == 0:
        break

    else:
        print('번호를 확인해주세요')
        continue
```

```python
# 네이버 영화 순위 불러오기 => df => csv 저장
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as req
#old_content > table > tbody > tr:nth-child(2) > td:nth-child(1) > img
#old_content > table > tbody > tr > td.title > div > a
url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver"

res = req.urlopen(url)
soup = BeautifulSoup(res,"html.parser")

# 순위
ranks = soup.select("#old_content > table > tbody > tr > td > img")
r_list = [r.attrs['alt'] for r in ranks]

# 순위에서 숫자만 추출 (변동폭 제외)
# new_r_list = []
# for r in r_list:
#     try:
#         new_r_list.append(int(r))
#     except:
#         pass
# print(new_r_list)

# 순위에 빈 칸이 있으므로 숫자 직접 지정
new_r_list = list(range(1,51))
print(new_r_list)

# 제목
titles = soup.select("td.title")
t_list = [t.text.replace("\n","") for t in titles]
print(t_list)

# Dataframe
df = pd.DataFrame(data=zip(new_r_list,t_list),columns=["Index","Title"])

df.to_csv("../Data/NaverList.csv",index=None)
pd.read_csv("../Data/NaverList.csv",index_col=0)
```
