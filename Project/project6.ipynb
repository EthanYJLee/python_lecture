```python
import pandas as pd
from glob import glob

file_names = glob("../JejuData/Jeju_Card_2017_2021/*.csv") # 폴더 내의 모든 csv파일 목록을 불러온다
card = pd.DataFrame() # 빈 데이터프레임 하나를 생성한다
for file_name in file_names:
    temp = pd.read_csv(file_name, encoding='euc-kr') # csv파일을 하나씩 열어 임시 데이터프레임으로 생성한다
    card = pd.concat([card, temp]) # 전체 데이터프레임에 추가하여 넣는다


card.sort_values('년월',ascending=True)
```
