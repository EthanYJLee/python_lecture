> 여러 파일들 (파일 이름들) 한 번에 불러오기 - glob

```python
from glob import glob
```

```python
station_files = glob("../Data/oil/*.xls") # file names를 list 형식으로 불러온다 <- for문으로 처리할 것
station_files
```

```python
tmp_raw = []
for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)
station_raw = pd.concat(tmp_raw)
station_raw.head(30)
```

```python
len(station_raw)
```

```python
station_raw.info()
```

```python
# 필요 항목 발췌
stations = pd.DataFrame(
    {
        'Oil_Store':station_raw['상호'],
        '주소':station_raw['주소'],
        '가격':station_raw['휘발유'],
        '셀프':station_raw['셀프여부'],
        '상표':station_raw['상표']
    }
)
stations.head()
stations
```

```python
stations.info()
```

```python
stations['구'] = [region.split()[1] for region in stations['주소']]
stations
```

```python
# 구 컬럼 확인
stations['구'].unique()
```

```python
# 여러개 파일을 합하여 만든 데이터셋이기 때문에 index가 중복되어 있음
stations.reset_index(inplace=True, drop=True)
stations
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc

# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 셀프와 셀프가 아닌 주유소의 가격 차이를 박스플롯으로 확인
# stations
plt.figure(figsize=(8,4))
plt.boxplot(
    x=[
        stations[stations['셀프'] == 'Y']['가격'],
        stations[stations['셀프'] == 'N']['가격']
    ],
    labels=stations.셀프.unique()
)
plt.title('Boxplot group by 셀프', size=15),
plt.xlabel('셀프')
plt.ylabel('휘발유 가격')
plt.grid()
plt.show()
```
