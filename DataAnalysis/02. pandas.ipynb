```python
data = pd.DataFrame({
    'c0':[1,2,3],
    'c1':[4,5,6],
    'c2':[7,8,9],
    'c3':[10,11,12],
    'c4':[13,14,15]
})
data.index = ['r0','r1','r2']
data
```

```python
# 인덱스 재설정
data.index
# r3, r4 index 추가하고
new_index = ['r0','r1','r2','r3','r4']
# NaN 값을 0으로 fill
ndf = data.reindex(new_index, fill_value=0)
ndf
```

```python
# index reset (기본 index 1,2,3,4... 를 새로운 index로 생성)
ndf.reset_index()
```

```python
# 인덱스 정렬하기
# 역순 정렬
data.sort_index(ascending=False)
# inplace = True 하면 저장
```

```python
# c1을 기준으로 내림차순 정렬 후 바로 적용하기
data.sort_values(by='c1',ascending=False, inplace=True)
data
```
