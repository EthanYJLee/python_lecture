```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
traffic_Seoul = pd.read_csv('../Data/newSeoul_2005_2019.csv')
traffic_Seoul
```

```python
# 2019년 데이터만 추출하기
traffic_Seoul_2019 = traffic_Seoul[traffic_Seoul['년도']==2019]
traffic_Seoul_2019
```

```python
# index 정리하기
traffic_Seoul_2019.reset_index(drop=True,inplace=True)
traffic_Seoul_2019
```

```python
# 자치구별 발생건수, 사망자수, 부상자수 집계하기
traffic_anal = pd.pivot_table(traffic_Seoul_2019,index='자치구명',aggfunc=sum)
traffic_anal
```

```python
# 년도와 월은 제거하여 Dataset 만들기
traffic_anal = traffic_anal[['발생건수','부상자수','사망자수']]
traffic_anal
```

```python
seoul_limit = pd.read_csv('../Data/seoul.csv',encoding='euc-kr')
seoul_limit
```

```python
seoul_limit.columns = ['자치구명','lon','lat']
seoul_limit
```

```python
data_result = pd.merge(traffic_anal,seoul_limit,on='자치구명')
data_result
```

```python
data_result.info()
```

```python
# folium을 사용하기 위해 int를 float로 변환
data_result.iloc[:,1:4] = data_result.iloc[:,1:4].astype('float')
data_result.info()
```

```python
import folium
from folium import plugins
```

```python
# 자치구의 구청을 표시하기
map = folium.Map(location=[37.55,126.98],zoom_start=11,width=1000,height=500)
for lat,lng,name,acc,inj,dea in zip(data_result['lat'],data_result['lon'],data_result['자치구명'],data_result['발생건수'],data_result['부상자수'],data_result['사망자수']):
    folium.Marker(location=[lat,lng],tooltip=[name,'\n발생건수:',acc,'\n사망자수:',dea]).add_to(map)
    folium.CircleMarker(location=[lat,lng],radius=acc/35, color='blue',fill_color='skyblue',fill=True,fill_opacity=0.7).add_to(map)
    folium.CircleMarker(location=[lat,lng],radius=dea*2,color='red',fill_color='YlOrRd',fill=True,fill_opacity=0.3).add_to(map)
minimap = plugins.MiniMap()
map.add_child(minimap)
map.add_child(plugins.MousePosition())
map.add_child(plugins.Draw())
map
```
