### Feature Engineering
- 모델 정확도를 높이기 위해서 주어진 데이터를 예측 모델의 문제를 잘 표현할 수 있는 features로 변형시키는 과정 (적분 등)
- FamilySize라는 Feature Column 생성

```python
# Embarked에 따른 컬럼을 생성
# Q, S => UK, C => Austria
# Nationality_AT, Nationality_UK라는 컬럼을 생성하고 True/False값 출력되도록 추가
train['Nationality_AT'] = (train.Embarked == 'C')
train['Nationality_UK'] = (train.Embarked != 'C')
train[['Embarked','Nationality_AT','Nationality_UK']]
```

```python
train.head()
```

```python
# Nationality 컬럼 생성 (England, Austria)
# nationality = []
# nat = list(train.Nationality_UK)
# for n in nat:
#     if(n == True):
#         nationality.append('England')
#     else:
#         nationality.append('Austria')


# train['Nationality'] = nationality
# train.head()
```

```python
train.loc[train['Embarked'] == 'C', 'Nationality'] = 'Austria'
train.loc[train['Embarked'] != 'C', 'Nationality'] = 'England'
train.head()
```

```python
# Fare에 대해 분류 컬럼을 만들려고 한다. 기준점은?
train.Fare.describe()
print('1사분위수: ', train.Fare.quantile(.25))
print('2사분위수: ', train.Fare.quantile(.5))
print('3사분위수: ', train.Fare.quantile(.75))
print('4사분위수: ', train.Fare.quantile(1))
# 주로 2, 3분위수로 3분할
```

```python
# 2분위수와 3분위수의 근사 정수치를 선택
train['LowFare'] = train['Fare'] < 15
train['MediumFare'] = (train['Fare'] >= 15) & (train['Fare'] < 30)
train['HighFare'] = train['Fare'] >= 30

train[['Fare','LowFare','MediumFare','HighFare']]
```

```python
# FareType이라는 컬럼 생성
train.loc[train['Fare'] < 15, 'FareType'] = 'Low'
train.loc[(train['Fare'] >= 15) & (train['Fare'] < 30), 'FareType'] = 'Medium'
train.loc[train['Fare'] >= 30, 'FareType'] = 'High'
train[['Fare','FareType']].head(10)
```

```python
# Python의 함수를 이용하여 처리하기
def check_fare_type(fare):
    if fare < 15:
        return 'Low'
    elif fare >= 15 and fare < 30:
        return 'Medium'
    else:
        return 'High'
```

```python
train['FareType2'] = train['Fare'].apply(check_fare_type)
# 함수를 부르는 함수 'apply'
train[['Fare','FareType','FareType2']]
```

```python
# Age의 NaN을 Age의 평균값으로 대체하기 (단, 소수점 첫째자리까지 반올림하기)
train.loc[train.Age.isnull()] = round(train.Age.mean(),1)
```
