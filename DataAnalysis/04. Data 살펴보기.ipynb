### Pandas의 내장 그래프 도구

```python
# 남북한의 발전량 데이터셋 불러오기
# df = pd.read_excel("../Data/남북한발전전력량.xlsx", engine='openpyxl')
df = pd.read_excel("../Data/남북한발전전력량.xlsx")
```

```python
df.head()
```

```python
df.tail()
```

```python
# 남한, 북한 발전량 합계 데이터만 추출하여 df_ns로 만들기
# df_ns의 행인덱스를 South와 North로 변경하기
df_ns = df.iloc[[0,5],2:]
df_ns.index = ['South','North']
df_ns
```

```python
df_ns.info()
```

```python
# object -> int로 변환
df_ns = df_ns.astype('int')
df_ns
```

```python
df_ns.info()
```

```python
# 선 그래프
df_ns.columns
```

```python
tdf_ns = df_ns.transpose()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10,5)
```

```python
tdf_ns.plot()
```

```python
# 막대 그래프
tdf_ns.plot(kind='bar')
```

```python
# 히스토그램
tdf_ns.plot(kind='hist')
```
