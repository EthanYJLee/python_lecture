### 2개의 y축을 갖는 그래프 작성
- 남북한 발전량
- 막대 그래프는 북한의 발전량, 선 그래프는 북한 발전량의 전년 대비 증감률을 백분율로 표시

```python
df = pd.read_excel("../Data/남북한발전전력량.xlsx", convert_float=True)
df
```

```python
# 북한 전력량만 선택
df = df.loc[5:9]
df.drop('전력량 (억㎾h)',axis=1,inplace=True)
df
```

```python
# 발전전력별을 index로 변경하기
df.set_index('발전 전력별',inplace=True)
df
```

```python
df
```

```python
# 기준점을 발전 전력별로 변경
df = df.T
df
```

```python
# 합게 컬럼을 총 발전량으로 변경
df = df.rename(columns={'합계':'총 발전량'})
df
# 전년도 발전량 컬럼을 생성해 증감량 데이터 계산
df['총발전량-1년'] = df['총 발전량'].shift(1) # shift: 전 (prev)행의 데이터를 삭제
df
```

```python
# 증감률 컬럼 생성
df['증감률'] = (df['총 발전량']/df['총발전량-1년']-1)*100
df
```

```python
plt.style.use('ggplot')
ax1 = df[['수력','화력']].plot(
    kind='bar',
    figsize=(20,10),
    stacked=True
)
ax2 = ax1.twinx()
ax2.plot(
    df.index,
    df.증감률,
    ls='--',
    marker='o',
    markersize=10,
    color='green',
    label='전년대비 증감률 (%)'
)
ax1.set_ylim(0,500)
ax2.set_ylim(-50,50)
ax1.set_xlabel('연도',size=20)
ax1.set_ylabel('발전량 (억 kWh')
ax2.set_ylabel('전년대비 증감률 (%)')
ax1.legend(loc='upper left',fontsize=15)
ax2.legend(loc='upper right',fontsize=15)

plt.title('북한 전력 발전량 (1990~2016)',size=30)

plt.show()
```

```python
# 자동차의 국가별 데이터 개수를 파이차트로 표현
# 1:US; 2:EU; 3:JP;
# 합계를 구하기 위해 전부 1인 컬럼 만들어주기
df['count'] = 1
df
```

```python
# origin 열을 기준으로 그룹화, 합계
df_origin = df.groupby('origin').sum()
df_origin
```

```python
df_origin.index = ['USA','EU','JAPAN']
df_origin
```

```python
df_origin['count'].plot(
    kind='pie',
    figsize=(7,5),
    autopct='%.2f%%',
    startangle=10,
    colors = ['chocolate','bisque','cadetblue']
)

plt.title('Model Origin',size=20)
plt.axis('equal')
plt.legend(
    labels=df_origin.index,
    loc='upper right'
    )
plt.show()
```

### Box Plot
- 범주형 데이터의 분포 확인

```python
plt.style.use('seaborn-poster')
df = pd.read_csv("../Data/auto-mpg.csv", header=None)
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
```

```python
# 차트 제목 추가
import platform
from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.boxplot(
    x=[
        df[df['origin']==1]['mpg'],
        df[df['origin']==2]['mpg'],
        df[df['origin']==3]['mpg'],
    ],
    labels=['USA','EU','JAPAN']
)
ax2.boxplot(
    x=[
        df[df['origin']==1]['mpg'],
        df[df['origin']==2]['mpg'],
        df[df['origin']==3]['mpg'],
    ],
    labels=['USA','EU','JAPAN'],
    vert=False
)
ax1.set_title('제조국가별 연비 분포 (수직 박스 플롯)')
ax1.set_title('제조국가별 연비 분포 (수평 박스 플롯)')
plt.show()
```
