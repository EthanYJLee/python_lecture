```python
number_of_data = 18
img_width = 128
img_height = 128
```

```python
import numpy as np
from PIL import Image
import glob
```

```python
train = np.zeros(number_of_data*img_height*img_width,dtype=np.int32).reshape(number_of_data,img_height,img_width)
i = 0
for file in sorted(glob.glob("./Data/FaceGray/Cathy/*.jpg")):
    img = Image.open(file).convert('L').resize((img_width,img_height))
    img = np.array(img,dtype=np.int32)
    train[i,:,:] = img
    i+=1
```

```python
import pandas as pd
train = train.reshape(-1,img_width*img_height)
df = pd.DataFrame(train)
df.head()
```
