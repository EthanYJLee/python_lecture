```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from tensorflow import keras
```

```python
data = pd.read_csv("./Data/gpascore.csv")
data.head()
# admit: 합격여부; gre: 영어성적; gpa: 학점; rank: 지원한 대학원 수준
```

```python
# 결측치 확인
data.isnull().sum()
# 결측치 데이터 삭제
data.dropna(inplace=True)

# rank의 종류
sorted(data['rank'].unique())
# [1, 2, 3, 4]

# 영어 점수의 최소값과 최대값
print('최대값:',data['gre'].max())
print('최소값:',data['gre'].min())
# 최대값: 800.0, 최소값: 220.0

# 합격여부 데이터 빈도수 확인
print('합격수:',data[data['admit']==1]['admit'].count())
print('불합격수:',data[data['admit']!=1]['admit'].count())
# 합격수: 201, 불합격수: 224
data[data['admit']==1]['admit'].count()/data[data['admit']!=1]['admit'].count()
# 0.8973214285714286
```

```python
# Features
x = data[['gre','gpa','rank']]
# Target
y = data['admit']
```

## Deep Learning Model 만들기
- 0 or 1이므로 Sigmoid함수를 사용한다.

```python
model = keras.Sequential()
model.add(keras.layers.Dense(32, activation='tanh',input_shape=(3,)))
model.add(keras.layers.Dense(256, activation='tanh',input_shape=(3,)))
model.add(keras.layers.Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
model.fit(np.array(x),np.array(y),epochs=500, verbose=1)
```

```python
model.evaluate(np.array(x),np.array(y))
```

```python
# 예측하기
# gre: 700; 학점: 3.7, Rank: 4인 학교
model.predict([[700,3.7,1],[700,3.7,2],[700,3.7,3],[700,3.7,4]])
```
