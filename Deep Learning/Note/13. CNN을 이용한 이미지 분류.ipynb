```python
import numpy as np
from tensorflow import keras
import glob
import matplotlib.pyplot as plt
from PIL import Image
```

```python
targets = {'0L':0,
           '1L':1,
           '2L':2,
           '3L':3,
           '4L':4,
           '5L':5,
           '0R':6,
           '1R':7,
           '2R':8,
           '3R':9,
           '4R':10,
           '5R':11,
           }
```

### train load

```python
X_train = []
train_labels = []
widthRatio = []
heightRatio = []

for file in sorted(glob.glob("./Data/fingers/train/*.png")):
    img = np.array(Image.open(file),dtype=np.int32)
    # target 숫자로 변경
    train_labels.append(targets[file[-6:-4]])
    widthRatio.append(img.shape[1])
    heightRatio.append(img.shape[0])
    X_train.append(img)

print('너비 평균 해상도:',np.mean(widthRatio))
print('높이 평균 해상도:',np.mean(heightRatio))
print(len(widthRatio))
print(len(heightRatio))
train_labels
```

### train resize

```python
i = 0
for file in sorted(glob.glob("./Data/fingers/train/*.png")):
    img = Image.open(file)
    # sampling: 해상도 확장 시 비어있는 부분을 채우는 방식
    imgResize = img.resize((32,32),Image.Resampling.LANCZOS)
    imgResize = imgResize.convert('L') # Gray Scale, 투명도가 있는 경우 LA
    imgResize.save("./Data/fingers/trainNew/image_%04d.png"%i, 'png')
    i += 1
    if i >= 18000:
        break
```

### test load

```python
X_test = []
test_labels = []
widthRatio = []
heightRatio = []
for file in sorted(glob.glob("./Data/fingers/test/*.png")):
    img = np.array(Image.open(file),dtype=np.int32)
    # target 숫자로 변경
    test_labels.append(targets[file[-6:-4]])
    widthRatio.append(img.shape[1])
    heightRatio.append(img.shape[0])
    X_test.append(img)

print('너비 평균 해상도:',np.mean(widthRatio))
print('높이 평균 해상도:',np.mean(heightRatio))
print(len(widthRatio))
print(len(heightRatio))
test_labels
```

### test resize

```python
i = 0
for file in sorted(glob.glob("./Data/fingers/test/*.png")):
    img = Image.open(file)
    # sampling: 해상도 확장 시 비어있는 부분을 채우는 방식
    imgResize = img.resize((32,32),Image.Resampling.LANCZOS)
    imgResize = imgResize.convert('L') # Gray Scale, 투명도가 있는 경우 LA
    imgResize.save("./Data/fingers/testNew/image_%04d.png"%i, 'png')
    i += 1
    if i >= 3600:
        break
```

### resized image load

```python
X_train = []
for file in sorted(glob.glob("./Data/fingers/trainNew/*.png")):
    img = np.array(Image.open(file),dtype=np.int32)
    widthRatio.append(img.shape[1])
    heightRatio.append(img.shape[0])
    X_train.append(img)

X_test = []
for file in sorted(glob.glob("./Data/fingers/testNew/*.png")):
    img = np.array(Image.open(file),dtype=np.int32)
    widthRatio.append(img.shape[1])
    heightRatio.append(img.shape[0])
    X_test.append(img)
```

```python
# train data 만들기
tempData = np.zeros(18000*32*32, dtype=np.int32).reshape(18000,32,32)

i = 0
for image in X_train:
    img = np.array(image, dtype=np.int32)
    tempData[i,:,:] = img
    i+=1
X_train = tempData.copy()
```

```python
# test data 만들기
tempData = np.zeros(3600*32*32, dtype=np.int32).reshape(3600,32,32)

i = 0
for image in X_test:
    img = np.array(image, dtype=np.int32)
    tempData[i,:,:] = img
    i+=1
X_test = tempData.copy()
```

```python
np.array(train_labels)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_scaled = X_train.reshape(-1, 32, 32, 1) / 255.0
train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, np.array(train_labels), test_size=0.2, random_state=42)
```

```python
print(train_scaled.shape)
print(train_target.shape)
print(val_scaled.shape)
print(val_target.shape)
```

```python
train_target
```

```python
# model
model = keras.Sequential()
# 합성곱 층
model.add(keras.layers.Conv2D(32, kernel_size = 3, activation = 'relu', padding='same', input_shape = (32,32,1)))
# 풀링 층
model.add(keras.layers.MaxPooling2D(2))
# model.add(keras.layers.AveragePooling2D(2))
model.add(keras.layers.Conv2D(64, kernel_size = 3, activation='relu', padding='same'))
model.add(keras.layers.MaxPooling2D(2))
# model.add(keras.layers.AveragePooling2D(2))
# Dense 층
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(100, activation='relu'))
# model.add(keras.layers.Dropout(0.4))
model.add(keras.layers.Dense(12,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics='accuracy')
checkpoint_cb = keras.callbacks.ModelCheckpoint("./Data/best_cnn_fingers.h5")
early_stopping_cb = keras.callbacks.EarlyStopping(patience=2,restore_best_weights=True)
history = model.fit(train_scaled, train_target, epochs=20, validation_data=[val_scaled,val_target],callbacks=[checkpoint_cb,early_stopping_cb])
```
