```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fruits = np.load("../Data/fruits.npy")
fruits.shape
```

```python
# Image 저장하기
from PIL import Image
```

```python
for i in range(300):
    image = Image.fromarray(fruits[i]).convert('L') # L: 흑백톤
    image.save("../Data/fruits/fruit_%03d.jpg"%i,"JPEG")
```

### image를 읽어서 Numpy로 변환하기

```python
import glob
```

```python
numberOfData = 300
imgSize = 100
```

```python
# 배열 초기화
fruitsData = \
    np.zeros(numberOfData*imgSize*imgSize,dtype=np.int32).reshape(numberOfData,imgSize,imgSize)
```

```python
i = 0
for file in sorted(glob.glob("../Data/fruits/*.jpg")):
    img = np.array(Image.open(file),dtype=np.int32)
    fruitsData[i,:,:] = img
    i+=1
```

```python
plt.imshow(fruitsData[0].reshape(100,100),cmap="gray_r")
plt.show()
```

```python
plt.imshow(fruitsData[100].reshape(100,100),cmap="gray_r")
plt.show()
```

```python
plt.imshow(fruitsData[200].reshape(100,100),cmap="gray_r")
plt.show()
```

```python
# 저장하기
np.save("../Data/fruitsData",fruitsData)
```
